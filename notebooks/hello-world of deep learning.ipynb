# Import statements

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

# Create a Nural network with One layer and one neuron
where input dimentions are scaler 

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

# We Built a model Now we have to train it 
so first we have to decide which Loss function and which optimizer we should use.

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

# Let's Define our traing set of 6 examples 

```python
x=np.array([1.0, 2.0,  3.0, 4.0, 6.0, 7.0], dtype=float)
y=np.array([4.0, 6.0,  8.0, 10.0, 14.0, 16.0], dtype=float)
```

# It's Time to fit our model
if we increase no of epochs it will effect our accuraccy in good manner.

```python
model.fit(x,y,epochs=999)
```

```python
print(model.predict([5.0]))
```
